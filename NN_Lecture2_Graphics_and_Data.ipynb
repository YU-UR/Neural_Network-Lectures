```python
import numpy as np
import matplotlib.pyplot as plt


def f1(x):
    return 4+x

def f2(x):
    f1(x)
    return f1(x)+0.5*x*x

def f3(x):
    f2(x)
    return f2(x)+0.2*x**3

x1 = np.arange(-1, 1, 0.2) #x-values
plt.plot(x1,f1(x1),'b', label='f1');
plt.plot(x1,f2(x1),'g', label='f2');
plt.plot(x1,f3(x1),'r', label='f3');
plt.title('function of f1,2,3')
plt.legend(loc='upper right');
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f1(x):
    return 4 + x

def f2(x):
    return 4 + x + 0.5 * x**2

def f3(x):
    return 4 + x + 0.5 * x**2 + 0.2 * x**3

def f4(x):
    return (f1(x) + f2(x) + f3(x)) / 3

# Create a range of x values for plotting
x = np.linspace(-5, 5, 400)

# Create a plot with subplots for each function
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot f1
axs[0, 0].plot(x, f1(x), label=r"$f_1(x) = 4 + x$", color='b')
axs[0, 0].set_title(r"$f_1(x)$")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("f(x)")
axs[0, 0].grid(True)
axs[0, 0].legend()

# Plot f2
axs[0, 1].plot(x, f2(x), label=r"$f_2(x) = 4 + x + 0.5x^2$", color='g')
axs[0, 1].set_title(r"$f_2(x)$")
axs[0, 1].set_xlabel("x")
axs[0, 1].set_ylabel("f(x)")
axs[0, 1].grid(True)
axs[0, 1].legend()

# Plot f3
axs[1, 0].plot(x, f3(x), label=r"$f_3(x) = 4 + x + 0.5x^2 + 0.2x^3$", color='r')
axs[1, 0].set_title(r"$f_3(x)$")
axs[1, 0].set_xlabel("x")
axs[1, 0].set_ylabel("f(x)")
axs[1, 0].grid(True)
axs[1, 0].legend()

# Plot f4
axs[1, 1].plot(x, f4(x), label=r"$f_4(x) = \frac{1}{3}(f_1(x) + f_2(x) + f_3(x))$", color='purple')
axs[1, 1].set_title(r"$f_4(x)$")
axs[1, 1].set_xlabel("x")
axs[1, 1].set_ylabel("f(x)")
axs[1, 1].grid(True)
axs[1, 1].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

### 2.3.2) Contour-Plots with $\textit{plt.contour()}$

Contour-Plots are a good tool to visualize functions $f(x,y)=z$ in a 2-D Graphics.  
For each tuple $(x,y)$ the value $f(x,y)=z$ is shown as a contour-line.  
This creates a kind of map of the function values.

```python
def f(x, y):
    return (1 - x/2+x**5+y**3)*np.exp(-x**2-y**2) # function f(x,y)=z

n = 256 #number of 
x = np.linspace(-3,3,n) # x-values
y = np.linspace(-3,3,n) # y-values


X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='jet')
C = plt.contour(X, Y, f(X, Y), 8, colors='black')

plt.clabel(C, inline=1, fontsize=10) 
plt.title('Default with labels');
```

### Exercise 3) 
- Define the functions $f_1(x,y) = sin(x) + cos(y)$ and $f_2(x,y) = sin(x) \cdot cos(y)$.
- Generate two Contour-Plots for $f_1(x)$ und $f_2(x)$ for x and y from $I=[-Pi,Pi]$. 
- Check the documentation (e.g., http://matplotlib.org/examples/index.html) for ways to style the graphics differently (color of the areas, color and thickness of the contours, labels, size, font type, etc.) and experiment with the design

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions f1 and f2
def f1(x, y):
    return np.sin(x) + np.cos(y)

def f2(x, y):
    return np.sin(x) * np.cos(y)

# Set up the grid for x and y in the interval [-pi, pi]
x = np.linspace(-np.pi, np.pi, 400)  # 400 points between -pi and pi
y = np.linspace(-np.pi, np.pi, 400)
X, Y = np.meshgrid(x, y)  # Create a meshgrid for the 2D plot

# Compute the function values
Z1 = f1(X, Y)
Z2 = f2(X, Y)

# Create the contour plot for f1(x, y)
plt.figure(figsize=(12, 6))

# First subplot for f1(x, y)
plt.subplot(1, 2, 1)
cp1 = plt.contourf(X, Y, Z1, 20, cmap='coolwarm', alpha=0.75)  # Filled contours with color map
contour1 = plt.contour(X, Y, Z1, 20, colors='black', linewidths=1)  # Contour lines
plt.clabel(contour1, inline=True, fontsize=10, fmt='%1.1f', colors='black')  # Add labels to contour lines
plt.title(r'$f_1(x, y) = \sin(x) + \cos(y)$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.colorbar(cp1, label='Function Value')  # Add color bar

# Second subplot for f2(x, y)
plt.subplot(1, 2, 2)
cp2 = plt.contourf(X, Y, Z2, 20, cmap='viridis', alpha=0.75)  # Different color map for f2
contour2 = plt.contour(X, Y, Z2, 20, colors='white', linewidths=1.5)  # Contour lines with white color
plt.clabel(contour2, inline=True, fontsize=10, fmt='%1.1f', colors='white')  # Add labels to contour lines
plt.title(r'$f_2(x, y) = \sin(x) \cdot \cos(y)$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.colorbar(cp2, label='Function Value')  # Add color bar

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function f1(x, y) = sin(x) + cos(y)
def f1(x, y):
    return np.sin(x) + np.cos(y)

# Set up the grid for x and y in the interval [-pi, pi]
x = np.linspace(-np.pi, np.pi, 400)  # 400 points between -pi and pi
y = np.linspace(-np.pi, np.pi, 400)
X, Y = np.meshgrid(x, y)  # Create a meshgrid for the 2D plot

# Compute the function values
Z = f1(X, Y)

# Create a figure
fig = plt.figure(figsize=(10, 8))

# Add 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='jet', edgecolor='none', alpha=0.7)

# Add color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label='Function Value')

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('f1(x, y)')

# Add title
ax.set_title(r'$f_1(x, y) = \sin(x) + \cos(y)$', fontsize=14)

# Show plot
plt.show()
```
